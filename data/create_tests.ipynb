```python
website = r"https://www.spanishdict.com/conjugate/el%20salto"
response = requests.get(website)
html_content = response.text

# Step 3: Parse the HTML content using html5lib
soup = BeautifulSoup(html_content, 'html5lib')

# Step 4: Use an appropriate selector to extract the desired element
target_div = soup.find('div', class_='OSnUqYPB WVvCPjEU')

if target_div:
    target_text = ''
    for child in target_div.children:
        if child.name == 'span' and child.get('class') == ['conj-irregular']:
            target_text += child.text
        else:
            target_text += child

    print("Extracted text:", target_text)
else:
    print("Element not found")
```

```python
pronouns = ["yo", "t√∫", "√©l/ella/usted", "nosotros", "vosotros", "ellos/ellas/ustedes"]
sp = ["1s", "2s", "3s", "1p", "2p", "3p"]
tense = ["PRES", "PAST", "IMP", "FUT"]
response_items = ["ordenar", "manejar", "pensar",
                  "terminar", "explicar", "suceder", "opar", "compartir", "re√≠rse", "contar", "ense√±ar", "entender"]
with open ("test_spanish_sentences.tsv", "w", encoding="utf-8") as f:
    for item in response_items:
        response = requests.get(f"https://www.spanishdict.com/conjugate/{item}")
        parsed_html = html.fromstring(response.content)

        for i, idx in enumerate([2,3,4,5,6,7]):
            for j, tidx in enumerate([2,3,4,6]):
                xp = f'//*[@id="conjugation-content-wrapper"]//tbody/tr[{idx}]/td[{tidx}]/div/div[1]/div'
                out = parsed_html.xpath(xp)
                if i == 0 and j == 0:
                    parsed_str = out[-6].text_content()
                else:
                    parsed_str = out[0].text_content()
                f.write(f"{sp[i]}+{tense[j]}+{item}\t{pronouns[i]} {parsed_str}\n")
                f.write(f"NEG+{sp[i]}+{tense[j]}+{item}\t{pronouns[i]} no {parsed_str}\n")
```
